```python
import sys
from pathlib import Path
import shutil
import pandas as pd
```

```python
path_file = r"IsaacLab\source\standalone\grasp_dataset.csv"
path_file_in = r"IsaacLab\source_project_main\grasp_dataset.csv"
file_path = Path(path_file)
file_path_in = Path(path_file_in)
if file_path.is_file():
   print(f"Файл найден: {file_path}")
   if file_path_in.is_file():
      print(f"Файл найден: {file_path_in}")
   else:
      shutil.copy(path_file, path_file_in)
else:
   sys.exit()
```

```python
df = df[df['grasp_force_N'] >= 0]
```

```python
num_negative = (df['grasp_force_N'] < 0).sum()
print(f"Число отрицательных строк: {num_negative}")
df[df['grasp_force_N'] < 0].head()
```

```python
sum_all = len(df)
print("\n" + "="*50)
print(f"Статистика исходов (всего экспериментов: {sum_all})")
print("="*50)
counts = df["outcome"].value_counts()
percents = df["outcome"].value_counts()/sum_all * 100
for outcome in ["success", "drop", "crush", "marginal"]:
    if outcome in counts.index:
        cnt = counts[outcome]
        pct = percents[outcome]
        print(f"{outcome:10} : {cnt:5d} шт  →  {pct:5.1f}%")
    else:
        print(f"{outcome:10} :     0 шт  →   0.0%")

print("-"*50)
print(f"Всего записей: {sum_all}")
print("="*50 + "\n")
```
