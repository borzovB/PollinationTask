```python
print("Блок 1: Импорт библиотек выполнен")

import sys              # работа с аргументами и системой
from pathlib import Path  # удобная работа с путями
import shutil          # операции с файлами и папками
import pandas as pd    # работа с таблицами (DataFrame)
```

```python
path_file = r"IsaacLab\source\standalone\grasp_dataset.csv"      # путь к исходному файлу
path_file_in = r"IsaacLab\source_project_main\grasp_dataset.csv" # путь назначения

file_path = Path(path_file)        # создаём объект пути (исходный файл)
file_path_in = Path(path_file_in)  # создаём объект пути (файл назначения)

if file_path.is_file():  # проверка: существует ли исходный файл
   print(f"Блок 2: Файл найден: {file_path}")

   if file_path_in.is_file():  # проверка: существует ли файл в месте назначения
      print(f"Файл уже существует: {file_path_in}")
   else:
      shutil.copy(path_file, path_file_in)  # копируем файл
      print("Файл скопирован")
else:
   print("Исходный файл не найден")
   sys.exit()  # завершение программы
```

```python
print("Блок 3: Файл успешно загружен в DataFrame")
df = pd.read_csv(path_file_in)  # читаем CSV файл в DataFrame
```

```python
print("Блок 9: Проверка отрицательных значений в grasp_force_N")

num_negative = (df['grasp_force_N'] < 0).sum()  # считаем строки с отрицательной силой
print(f"Число отрицательных строк: {num_negative}")

print("Примеры строк с отрицательными значениями:")
df[df['grasp_force_N'] < 0].head()  # вывод первых 5 проблемных строк
```

```python
print("Блок 10: Удаление строк с отрицательной grasp_force_N")

df = df[df['grasp_force_N'] >= 0]  # оставляем только корректные значения

print("Фильтрация завершена: отрицательные значения удалены")
```

```python
print("Блок 11: Проверка отрицательных значений после очистки")

num_negative = (df['grasp_force_N'] < 0).sum()  # повторная проверка
print(f"Число отрицательных строк: {num_negative}")

print("Первые строки с отрицательными значениями (если есть):")
print(df[df['grasp_force_N'] < 0].head())
```

```python
sum_all = len(df)  # общее количество строк (экспериментов)

print("\n" + "="*50)
print(f"Блок 12: Статистика исходов (всего экспериментов: {sum_all})")
print("="*50)

counts = df["outcome"].value_counts()  # количество каждого исхода
percents = df["outcome"].value_counts() / sum_all * 100  # проценты

for outcome in ["success", "drop", "crush", "marginal"]:
    if outcome in counts.index:
        cnt = counts[outcome]
        pct = percents[outcome]
        print(f"{outcome:10} : {cnt:5d} шт  →  {pct:5.1f}%")
    else:
        print(f"{outcome:10} :     0 шт  →   0.0%")

print("-"*50)
print(f"Всего записей: {sum_all}")
print("="*50 + "\n")
```
