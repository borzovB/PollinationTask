```python
import os
import pandas as pd

# Загрузка набора данных
OUTPUT_FILE = os.path.join(os.getcwd(), "IsaacLab", "source", "standalone", "grasp_dataset.csv")

db = pd.read_csv(OUTPUT_FILE)
# Вывод данных
print(db.head(300))
```

```python
# Информация по столбцам
db.info()
```

```python
# Аналитика по столбцам
db.describe()
```

```python
if "outcome" not in db.columns:
    print("В файле отсутствует колонка 'outcome'")
    exit(1)

# ──────────────── Статистика ────────────────
counts = db["outcome"].value_counts()
percents = db["outcome"].value_counts(normalize=True) * 100
total = len(db)

print("\n" + "="*50)
print(f"Статистика исходов (всего экспериментов: {total})")
print("="*50)

for outcome in ["success", "drop", "crush", "marginal"]:
    if outcome in counts.index:
        cnt = counts[outcome]
        pct = percents[outcome]
        print(f"{outcome:10} : {cnt:5d} шт  →  {pct:5.1f}%")
    else:
        print(f"{outcome:10} :     0 шт  →   0.0%")

print("-"*50)
print(f"Всего записей: {total}")
print("="*50 + "\n")
```

```python
# выбрать только нужные колонки
columns = ["exp_id", "object_name", "grasp_force_N", "outcome"]
db_selected = db[columns]

print(db_selected.head(300))
```

```python
# Выберите нужные колонки
columns = ["exp_id", "object_name", "grasp_force_N", "outcome"]
df_selected = db[columns]

# Статистика по каждому объекту
stats = df_selected.groupby("object_name")["outcome"].value_counts().unstack(fill_value=0)
print(stats)

# Добавим общее количество экспериментов для каждого объекта
stats["total"] = stats.sum(axis=1)
print(stats)

# Процентное соотношение
stats_pct = stats.div(stats["total"], axis=0) * 100
print(stats_pct.round(1))
```
